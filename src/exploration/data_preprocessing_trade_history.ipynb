```python
import pandas as pd
import datetime
```

```python
def dateparse(datetime_str):    
    return datetime.datetime.strptime(datetime_str, '%Y-%m-%d %H:%M:%S')
```

```python
# Load dataset
ts = pd.read_csv('../../bld/out/data_raw/BTC_POT_trade_history.csv', parse_dates=True, index_col='date', date_parser=dateparse)

# Drop possible duplicates
ts = ts.loc[~ ts.tradeID.duplicated()]

# Drop unnecessary columns
ts.drop(['globalTradeID', 'tradeID'], axis=1, inplace=True)

# Save as pickle
ts.to_pickle('test.p')
```
